# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['python', 'c', 'c++', 'java']
y=[85,70,60,82]
z=[10,20,30,40]

width = 0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('languages')
plt.ylabel('Popularity')
plt.title('Bar Graph', fontsize=15)
color=['y','b', 'r','g']

plt.barh(
    p,y, 
    width, 
    color=color,
#     align='center', 
    edgecolor='r',
    linewidth=2,
    label='languages popularity'
)
plt.barh(
    p1,z, 
    width, 
    color='r',
#     align='center', 
    edgecolor='r',
    linewidth=2,
    label='languages popularity'
)

plt.yticks(p,x)
plt.legend()
plt.show()
```

# Scatter Plot

```python
x=[1,2,3,4]
y=[10,20,30,40]
colors=[22,15,36,14]

plt.scatter(x,y, c=colors, cmap='viridis', s=50)
t=plt.colorbar()
t.set_label('ColorBar')
plt.xlabel('Days')
plt.ylabel('Number')
plt.title('Scatter Plot')
plt.show()
```

# Pie chart

```python
x=[10,20,30,40]
y=['fuel', 'items', 'electricity', 'emi']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x, labels=y, 
        explode=ex, 
        colors=c, 
        autopct='%0.2f%%', 
        shadow=True, 
        radius= 0.5, 
        labeldistance=1.1, 
        startangle=90, 
        textprops={'fontsize':15}, 
        counterclock=False, 
        wedgeprops={'linewidth':4,'edgecolor':'m'}, 
        rotatelabels=False)
plt.title('Pie Chart')
plt.legend(loc=2)
plt.show()
```

# Bar Plot 

```python
x=[10,20,30,40,50,60,70,122]
plt.boxplot(
    x,
    vert=True,
    labels=['python'],
    showmeans=True,
    whis=1.5,
    sym='g_',
    boxprops=dict(color='r'),
    capprops=dict(color='g'),
    whiskerprops=dict(color='m'),
    flierprops=dict(markeredgecolor='y')
)
plt.show()
```

# Stack plot

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]
area2=[2,5,4,3,6]
area3=[6,5,3,4,1]
l=['area', 'area2', 'area3']

plt.stackplot(x,area,area2,area3, labels=l)
plt.legend(loc=2)
plt.show()
```

# Step Plot

```python
x=[1,2,3,4,5]
y=[2,4,3,5,1]

plt.step(
    x,y,
    marker='o',
    color='r',
    ms=10,
    mfc='g',
    label='python',
    
)

plt.title("Step plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=2)
plt.grid()
plt.show()
```

# Fill between plot

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y)
# plt.fill_between(x=[2,4],y1=2, y2=4)
plt.fill_between(x,y,where=(x>=2) & (x<=4), alpha=0.3)

plt.show()
```
